# CubesViewer from SQL database Example: Soccer Match DB

```python
import os
import sys

sys.path.append("../")
import cubesext

%load_ext autoreload
%autoreload 2
```

```python
db_url = 'sqlite:///data/match.sqlite3'
```

```python
(model_path) = cubesext.sql2cubes(db_url, model_path="generated/model-match.json", debug=False)
```

```python
# Launch a Cubes Server for this database and the generated model
print("DB URL: %s" % db_url)
print("Cubes Model path: %s" % model_path)
cubesext.cubes_serve(db_url, model_path, debug=True)
```

## Jupyter Notebook embedded CubesViewer view
